```python
from flask import Flask, request, jsonify
from flask_ngrok import run_with_ngrok
import torch
import cv2
import base64
import numpy as np
from yolov5 import detect

app = Flask(__name__)
run_with_ngrok(app)

model = torch.hub.load('ultralytics/yolov5', 'yolov5s')

def process_image(image):
    # Convert image to array and run YOLO
    results = model(image)
    results.render()  # Draw bounding boxes on the image
    img = results.imgs[0]
    _, img_encoded = cv2.imencode('.jpg', img)
    return base64.b64encode(img_encoded).decode('utf-8')

@app.route('/upload', methods=['POST'])
def upload_file():
    file = request.files['file']
    file_type = 'image'
    nparr = np.fromstring(file.read(), np.uint8)
    img = cv2.imdecode(nparr, cv2.IMREAD_COLOR)

    processed_image = process_image(img)
    return jsonify({"result": processed_image, "fileType": file_type})

if __name__ == '__main__':
    app.run()
```

```python
from IPython.display import HTML
import io

def detect_objects_in_video(video_path):
    # Lire la vidéo
    cap = cv2.VideoCapture(video_path)

    # Obtenir les informations de la vidéo
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))

    # Définir le codec et créer un objet VideoWriter
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    out = cv2.VideoWriter('output.avi', fourcc, fps, (width, height))

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Détection des objets
        results = model(frame)
        # Afficher les résultats sur l'image
        annotated_frame = results.render()[0]

        # Écrire le cadre annoté dans la vidéo de sortie
        out.write(annotated_frame)

        # Afficher le cadre annoté
        display(Image(data=cv2.imencode('.jpg', annotated_frame)[1].tobytes()))

    cap.release()
    out.release()
    cv2.destroyAllWindows()

# Testez avec une vidéo locale ou une URL de vidéo
video_path = 'https://ultralytics.com/images/zidane.jpg'  # Remplacez ceci par le chemin ou l'URL de votre vidéo
detect_objects_in_video(video_path)
```
